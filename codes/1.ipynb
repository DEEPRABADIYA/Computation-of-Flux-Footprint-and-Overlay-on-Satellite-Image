```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import footprint_FKM_climatology as fkm
```

```python
df = pd.read_csv('stable.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df1 = df[(df['mm'] == 12) | (df['mm'] < 3)]  # winter
df2 = df[(df['mm'] >=3) & (df['mm'] <= 5)]   # summer
df3 = df[(df['mm'] >= 6) & (df['mm'] <= 9)]   # monsoon
df4 = df[(df['mm'] >= 10) & (df['mm'] <= 11)]   # autumn
```

```python
df4['mm'].unique()
```

```python
ffp = fkm.FKM_climatology(df1['Zm_1'].tolist(), 
                                  df1['z0'].tolist(), 
                                  df1['umean'].tolist(),
                                  df1['ol'].tolist(), 
                                  df1['sigma_v'].tolist(), 
                                  df1['u_star'].tolist(), 
                                  df1['wind_dir'].tolist())
```

```python
ffp2 = fkm.FKM_climatology(df2['Zm_1'].tolist(), 
                                  df2['z0'].tolist(), 
                                  df2['umean'].tolist(),
                                  df2['ol'].tolist(), 
                                  df2['sigma_v'].tolist(), 
                                  df2['u_star'].tolist(), 
                                  df2['wind_dir'].tolist())
```

```python
ffp3 = fkm.FKM_climatology(df3['Zm_1'].tolist(), 
                                  df3['z0'].tolist(), 
                                  df3['umean'].tolist(),
                                  df3['ol'].tolist(), 
                                  df3['sigma_v'].tolist(), 
                                  df3['u_star'].tolist(), 
                                  df3['wind_dir'].tolist())
```

```python
ffp4 = fkm.FKM_climatology(df4['Zm_1'].tolist(), 
                                  df4['z0'].tolist(), 
                                  df4['umean'].tolist(),
                                  df4['ol'].tolist(), 
                                  df4['sigma_v'].tolist(), 
                                  df4['u_star'].tolist(), 
                                  df4['wind_dir'].tolist())
```

```python
# month 12 to 2
footprint_data = ffp['fclim_2d']
# Create a grid for the contour plot
x = np.linspace(-1500, 1500, footprint_data.shape[1])  # X-axis range
y = np.linspace(-1500, 1500, footprint_data.shape[0])  # Y-axis range
X, Y = np.meshgrid(x, y)

# Create a contour plot
plt.figure(figsize=(5, 6))
contour = plt.contourf(X, Y, footprint_data, levels=20, cmap='viridis')

# Add a colorbar
plt.colorbar(contour, label='Footprint Value')

# Add labels and title
plt.title('Zoomed Contour Plot of Footprint')
plt.xlabel('X-axis (m)')
plt.ylabel('Y-axis (m)')

# Set zoom limits (for example, zooming into the central region)
plt.xlim(-100, 100)  # Adjust the x-axis range for zoom
plt.ylim(-100, 100)  # Adjust the y-axis range for zoom

# Show the plot
plt.show()
```
