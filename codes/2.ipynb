```python
import pandas as pd
import footprint_FKM_climatology as fkm
```

```python
df = pd.read_csv('stable.csv')
df.head()
```

```python
df1 = df[(df['mm'] == 12) | (df['mm'] < 3)]  # winter
df2 = df[(df['mm'] >=3) & (df['mm'] <= 5)]   # summer
df3 = df[(df['mm'] >= 6) & (df['mm'] <= 9)]   # monsoon
df4 = df[(df['mm'] >= 10) & (df['mm'] <= 11)]   # autumn
```

```python
ffp = fkm.FKM_climatology(df1['Zm_2'].tolist(), 
                                  df1['z0'].tolist(), 
                                  df1['umean'].tolist(),
                                  df1['ol'].tolist(), 
                                  df1['sigma_v'].tolist(), 
                                  df1['u_star'].tolist(), 
                                  df1['wind_dir'].tolist())
```

```python
ffp2 = fkm.FKM_climatology(df2['Zm_2'].tolist(), 
                                  df2['z0'].tolist(), 
                                  df2['umean'].tolist(),
                                  df2['ol'].tolist(), 
                                  df2['sigma_v'].tolist(), 
                                  df2['u_star'].tolist(), 
                                  df2['wind_dir'].tolist())
```

```python
ffp3 = fkm.FKM_climatology(df3['Zm_2'].tolist(), 
                                  df3['z0'].tolist(), 
                                  df3['umean'].tolist(),
                                  df3['ol'].tolist(), 
                                  df3['sigma_v'].tolist(), 
                                  df3['u_star'].tolist(), 
                                  df3['wind_dir'].tolist())
```

```python
ffp4 = fkm.FKM_climatology(df4['Zm_2'].tolist(), 
                                  df4['z0'].tolist(), 
                                  df4['umean'].tolist(),
                                  df4['ol'].tolist(), 
                                  df4['sigma_v'].tolist(), 
                                  df4['u_star'].tolist(), 
                                  df4['wind_dir'].tolist())
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# month 3 to 5
footprint_data = ffp['fclim_2d']
# Create a grid for the contour plot
x = np.linspace(-1500, 1500, footprint_data.shape[1])  # X-axis range
y = np.linspace(-1500, 1500, footprint_data.shape[0])  # Y-axis range
X, Y = np.meshgrid(x, y)

# Create a contour plot
plt.figure(figsize=(8, 6))
contour = plt.contourf(X, Y, footprint_data, levels=20, cmap='viridis')

# Add a colorbar
plt.colorbar(contour, label='Footprint Value')

# Add labels and title
plt.title('Zoomed Contour Plot of Footprint')
plt.xlabel('X-axis (m)')
plt.ylabel('Y-axis (m)')

# Set zoom limits (for example, zooming into the central region)
plt.xlim(-100, 100)  # Adjust the x-axis range for zoom
plt.ylim(-100, 100)  # Adjust the y-axis range for zoom

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread

# Load your field image
image_path = 'new4.tif'  # Path to the uploaded image
field_image = imread(image_path)

# Month 12 to 2 footprint data
footprint_data = ffp['fclim_2d']

# Create a grid for the contour plot
x = np.linspace(-1300, 1300, footprint_data.shape[1])  # X-axis range
y = np.linspace(-1300, 1300, footprint_data.shape[0])  # Y-axis range
X, Y = np.meshgrid(x, y)

# Plot the field image in the background
plt.figure(figsize=(8, 6))

# Adjust the extent so that the image centers on the blue circle
# You'll need to adjust these numbers until the footprint aligns with the blue circle
plt.imshow(field_image, extent=[-100, 100, -100, 100], origin='upper', aspect='auto')
plt.scatter(lon_tower, lat_tower, color='black', marker='x', label='Flux Tower')


# Overlay the contour lines with a colormap
contour = plt.contour(X, Y, footprint_data, levels=20, cmap='viridis')

# Add a colorbar
plt.colorbar(contour, label='Footprint Value')

# Add labels and title
plt.title('Zoomed Contour Plot of Footprint with Field Image')
plt.xlabel('X-axis (m)')
plt.ylabel('Y-axis (m)')

# Set zoom limits
plt.xlim(-100, 100)  # Adjust the x-axis range for zoom
plt.ylim(-100, 100)  # Adjust the y-axis range for zoom

# Show the plot
plt.show()
```
